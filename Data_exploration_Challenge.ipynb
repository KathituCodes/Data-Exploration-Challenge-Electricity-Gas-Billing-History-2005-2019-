# Inspect the dataset for potential missing values

 Select your strategy to handle missing values,
 and tell us why you had made that choice

```python
#finding the total missing data in a column
data.isnull().sum()
```

```python
data["reading_remarque"].isnull().sum()
```

```python
#resolving the missing data in 'reading_remarque' by replacing with the mean, cause the data seem to have a pattern
data["reading_remarque"].fillna(data["reading_remarque"].mean(), inplace=True)
```

```python
data["reading_remarque"].isnull().sum()
```

```python
data.isnull().sum()
```

```python
data["counter_number"]
```

```python
data['counter_number'].fillna(method="ffill", inplace=True)
```

```python
data['counter_number'].isnull().sum()
```
